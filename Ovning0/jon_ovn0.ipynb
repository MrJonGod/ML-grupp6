```python
# Import packages
import feedparser
```

```python
# Define RSS feed URLs
RSS_URLS = [
    'https://www.dn.se/rss/',
    'https://www.svt.se/nyheter/lokalt/skane/rss.xml',
    'https://polisen.se/aktuellt/rss/skane/nyheter-rss---skane/'
]
```

```python

# Create an empty list for posts
posts = []
```

```python

# Loop through each RSS feed URL and extend the posts list with entries
for url in RSS_URLS:
    posts.extend(feedparser.parse(url).entries)
```

```python
# Loop through each RSS feed URL and append the posts list with entries
for url in RSS_URLS:
    posts.append(feedparser.parse(url).entries)
```

```python

# Print the collected articles (optional for debugging)
for post in posts:
    print(post)
```
